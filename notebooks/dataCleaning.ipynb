Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore", UserWarning)
```

Ingestion

```python
path = r'https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-03-churn-prediction/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df_telco = pd.read_csv(path)
df = df_telco.copy()
df.head()
```

Raw Data

```python
df.to_csv('../data/raw/rawdata.csv')
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
categorical_features = list(df.columns[df.dtypes == 'object'])
```

```python
for col in categorical_features:
    df[col] = df[col].str.lower().str.replace(' ','_')
```

```python
df.head()
```

```python
df['churn'] = (df['churn'] == 'yes').astype(int)
```

Processed Data

```python
df.to_csv(r'../data/processed/cleaned_data.csv')
```
